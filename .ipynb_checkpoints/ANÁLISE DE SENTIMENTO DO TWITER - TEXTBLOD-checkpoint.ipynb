```python
# Limpar os RT-Tweets do meu texto com comando find() a partir de posi√ß√£o

base = []

for i in dados:
    texto = (i.text)
    if('RT' in texto):
        posicao = texto.find(':')
        texto = texto[posicao+2:]
        base.append(texto)   
        
print(base)
```

```python
base_df = pd.DataFrame(base)
base_df.columns = ['text']
display(base_df)
```

```python
# Fun√ß√£o de limpeza de dados

def Limpeza_dados(text):
    text = re.sub(r'https?:\/\/\S+', '', text).lower().replace('.','').replace(';','').replace('-','').replace(':','').replace('()','').replace('@','').replace('#','').replace('!','').replace('$','').replace('%','').replace('+','').replace('(','').replace(')','').replace('[','').replace(']','').replace('/','').replace('?','')    
    base_df.drop_duplicates(['text'], inplace=True)
    return text

def RemoveStopWords(text):
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in text.split() if not i in stopwords]
    return (" ".join(palavras))

base_df['text'] = base_df['text'].apply(Limpeza_dados)
base_df['text'] = base_df['text'].apply(RemoveStopWords)
```

```python
base_df.head()
#len(base_df)
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

#criando novas colunas
base_df['Subjectivity']=base_df['text'].apply(getSubjectivity)
base_df['Polarity']=base_df['text'].apply(getPolarity) 

base_df.head()
```

```python
def getSentimento(score):
    if score < 0:
        return 'Negativo'
    elif score == 0:
        return 'Neutro'
    else:
        return 'Positivo'   
    
base_df['Sentimento'] = base_df['Polarity'].apply(getSentimento)

base_df.head()
```

```python
base_df['Sentimento'].value_counts()

plt.title('An√°lise de Sentimento')
plt.xlabel('Sentiment')
plt.ylabel('Counts')

base_df['Sentimento'].value_counts().plot(kind='bar')
plt.show()
```
