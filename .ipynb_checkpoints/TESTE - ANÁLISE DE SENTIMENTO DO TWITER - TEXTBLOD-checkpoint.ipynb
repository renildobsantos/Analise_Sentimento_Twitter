```python
# Removendo linhas duplicadas
texto_tweets.drop_duplicates(['text'], inplace=True)
texto_tweets.head()
#texto_tweets.count()
```

```python
# Fun√ß√£o para remover Stopwords da nossa base de dados:

def RemoveStopWords(instancia):
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in instancia.split() if not i in stopwords]
    return (" ".join(palavras))
```

```python
# Fun√ß√£o de limpeza de dados

def Limpeza_dados(instancia):
    instancia = re.sub(r"http\S+", "", instancia).lower().replace('.','').replace(';','').replace('-','').replace(':','').replace('()','').replace('@','').replace('#','').replace('!','').replace('$','')       
    return(instancia)
```

### Aplicando as 2 fun√ß√µes de Pr√©-processamento

```python
def Preprocessing(instancia):
    instancia = re.sub(r"http\S+", "", instancia).lower().replace('.','').replace(';','').replace('-','').replace(':','').replace('()','').replace('@','').replace('#','').replace('!','').replace('$','')
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in instancia.split() if not i in stopwords]
    return (" ".join(palavras))
```

```python
lista_tweets = []

for i in lista:
    x = Preprocessing(i)
    TextBlob(i).sentiment.subjectivity
    TextBlob(i).sentiment.polarity
    lista_tweets.append(x)
    
nova_df = pd.DataFrame(lista_tweets)
nova_df.columns = ['text']
display(nova_df)
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

def getSentimento(score):
    if score < 0:
        return 'Negativo'
    elif score == 0:
        return 'Neutro'
    else:
        return 'Positivo'
    
#criando novas colunas
nova_df['Subjectivity']=nova_df['text'].apply(getSubjectivity)
nova_df['Polarity']=nova_df['text'].apply(getPolarity)
nova_df['Sentimento'] = nova_df['Polarity'].apply(getSentimento)

nova_df
```

```python
nova_df['Sentimento'].value_counts()

plt.title('An√°lise de Sentimento')
plt.xlabel('Sentiment')
plt.ylabel('Counts')

nova_df['Sentimento'].value_counts().plot(kind='bar')
plt.show()
```
